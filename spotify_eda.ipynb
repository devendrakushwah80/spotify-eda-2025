```python
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
df['artist_name'].isnull().sum()
```

```python
df['artist_name'] = df['artist_name'].fillna("Unknown")
```

```python
df['artist_name'] = df['artist_name'].fillna("Unknown")
df['artist_genres'] = df['artist_genres'].fillna("Unknown")
```

```python
df['artist_name'] = df['artist_name'].astype(str)
df['artist_name'] = df['artist_name'].str.replace(r'[\$\{\}\(\)\[\]\*]', '', regex=True)
```

```python
top_artists = (
    df.groupby('artist_name')['artist_popularity']
      .mean()
      .sort_values(ascending=False)
      .head(20)
)
```

```python
plt.figure(figsize=(12,6))
plt.bar(top_artists.index, top_artists.values)
plt.xticks(rotation=90)
plt.xlabel("Artist Name")
plt.ylabel("Average Popularity")
plt.title("Top 20 Most Popular Artists")
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.histplot(df['track_popularity'], bins=30, kde=True, color='blue')
plt.title('Distribution of Track Popularity')       
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
top_artists = (
    df.groupby('artist_name')['track_number']
      .mean()
      .sort_values(ascending=False)
      .head(20)
)
```

```python

plt.figure(figsize=(12,6))
plt.bar(top_artists.index, top_artists.values)
plt.xlabel("Artist Name")
plt.ylabel("Average Track Number")
plt.title("Top 20 Artists by Average Track Number") 
plt.xticks(rotation=90)
plt.show()
```

```python
top_artists = (
    df.groupby('artist_name')['artist_followers']
      .mean()
      .sort_values(ascending=False)
      .head(20)
)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_artists.index, top_artists.values)
plt.xticks(rotation=90)
plt.xlabel("Artist Name")
plt.ylabel("Artist Followers")
plt.title("Top 20 Artists by Followers")
plt.show()
```
